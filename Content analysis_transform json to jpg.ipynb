```python
# Transforming Json to jpg - whole file and also craete meme_id (Conservative to CON000)
import os
import json
import requests

# Function to download images from JSON data
def download_images_from_json_file(json_file, output_dir, image_count):
    # Open the JSON file
    with open(json_file, 'r') as file:
        # Iterate over each line (assuming each line contains a JSON object)
        for line in file:
            try:
                # Load JSON data from the line
                data = json.loads(line)
                # Extract image URL
                image_url = data['url']
                # Download image
                response = requests.get(image_url)
                if response.status_code == 200:
                    # Generate image name with sequential code
                    image_name = f'LIB{image_count:03d}.jpg'  # Example: CON001.jpg, CON002.jpg, ...
                    image_count += 1
                    # Save image to file
                    image_path = os.path.join(output_dir, image_name)
                    with open(image_path, 'wb') as image_file:
                        image_file.write(response.content)
                        print(f"Downloaded: {image_name}")
                else:
                    print(f"Failed to download: {image_url}")
            except json.JSONDecodeError:
                print("Error decoding JSON")
    return image_count

# Path to the directory containing JSON files
json_dir = '/Users/amydoerr/Downloads/cameme/polset'  # Update with the path to your directory
# Output directory to save images
output_dir = '/Users/amydoerr/Desktop/memes'  # Update with the desired output directory

# Counter for image naming
image_count = 1

# Iterate over all JSON files in the directory
for filename in os.listdir(json_dir):
    if filename.endswith('.json'):
        json_file = os.path.join(json_dir, filename)
        image_count = download_images_from_json_file(json_file, output_dir, image_count)
```
