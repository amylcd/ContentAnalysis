```python
import pandas as pd

# URL of the CSV file
csv_url = '/Users/amydoerr/Downloads/Correct_full_df_CA.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_url)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming merged_df is your DataFrame
# Calculate the mean for 'I1.1' and 'I1.2' for each 'Lib_Con' value
grouped_means = df.groupby('Lib_Con')[['Mean_textualtone', 'Mean_visualtone']].mean()

# Calculate the overall mean of 'I1.1' and 'I1.2'
grouped_means['Mean_tone'] = df.groupby('Lib_Con')[['Mean_textualtone', 'Mean_visualtone']].mean().mean(axis=1)
```

```python
# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Define the labels and the positions of the bars
labels = ['Textual Tone', 'Visual Tone', 'combined Tone']
lib_con_0_means = grouped_means.loc[0]
lib_con_1_means = grouped_means.loc[1]
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

# Plot bars for Lib_Con == 0 in blue
ax.bar(x - width/2, lib_con_0_means, width, label='Liberal', color='blue')

# Plot bars for Lib_Con == 1 in red
ax.bar(x + width/2, lib_con_1_means, width, label='Conservative', color='red')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('Means')
ax.set_title('Means of Textual Tone, Visual Tone, and combined Tone')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Show the plot
plt.show()
```
