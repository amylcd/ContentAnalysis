```python
import os
import json
from datetime import datetime, timezone

# Define the directory containing the JSON files
directory = '/Users/amydoerr/Downloads/cameme/ConservativeMemes'
# Define the directory where you want to move the files that fall into the timeframe
new_directory = '/Users/amydoerr/Downloads/trytrytry'

# Ensure the new directory exists, create it if not
if not os.path.exists(new_directory):
    os.makedirs(new_directory)

# Get the current year
current_year = datetime.now().year

# Iterate over each JSON file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        with open(os.path.join(directory, filename), 'r') as file:
            data = json.load(file)
            # Extract the creation date from the JSON data
            creation_date = datetime.utcfromtimestamp(data['created_utc']).replace(tzinfo=timezone.utc)
            # Check if the creation date is within the last year
            if creation_date.year == current_year - 1:
                # Move the file to the new directory
                os.rename(os.path.join(directory, filename), os.path.join(new_directory, filename))
                print(f"Moved {filename} to {new_directory}")
            else:
                print(f"Skipping {filename}")
```
